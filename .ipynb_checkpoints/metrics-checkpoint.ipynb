```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X, y = make_classification(n_classes=2, n_samples=123456, n_informative=4)

X_train, X_test, y_train, y_test = train_test_split(X,y)

lr =LogisticRegression()
```

```python
from skore import EstimatorReport
```

```python
# 1. Train the model first!
lr.fit(X_train, y_train)

# 2. Initialize the reporter with the FITTED model
# (Fixed the typo 'X_trai' to 'X_train' here)
reporter = EstimatorReport(
    lr, 
    X_train=X_train, 
    y_train=y_train, 
    X_test=X_test, 
    y_test=y_test
)

# 3. Display the metrics
reporter.metrics.summarize().frame()
```

```python
reporter.help()
```

```python
reporter.metrics.confusion_matrix().plot()
```

```python
reporter.metrics.accuracy()
```
